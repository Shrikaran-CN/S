```python
import nbformat
from nbconvert import MarkdownExporter
from IPython.display import Markdown

def open_notebook(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        notebook = nbformat.read(f, as_version=4)
    return notebook

def notebook_to_markdown(notebook):
    markdown_exporter = MarkdownExporter()
    markdown, _ = markdown_exporter.from_notebook_node(notebook)
    return markdown

def display_markdown(markdown):
    display(Markdown(markdown))

while True:
    HWN = int(input("Enter homework number from the below menu \n\n1: BMI\n2: While Question String\n3: FindRemainder Function\n4: Lists with 0\n5: First Second File to Third\n6: Dictionary List\n7: Pandas Basics\n8: Stats & Pandas\n9: Data Visualization\n10: Address Class \n\nPress 0 to exit: "))    
    if HWN == 0:
        print("Exiting...")
        break
    elif HWN >= 1 and HWN <= 10:
        if HWN == 1:
            notebook_path = 'BUSA523_Shrikaran_HW_1.ipynb'
        elif HWN == 2:
            notebook_path = 'BUSA523_Shrikaran_HW_2.ipynb'
        elif HWN == 3:
            notebook_path = 'BUSA523_Shrikaran_HW_3.ipynb'
        elif HWN == 4:
            notebook_path = 'BUSA523_Shrikaran_HW_4.ipynb'
        elif HWN == 5:
            notebook_path = 'BUSA523_Shrikaran_HW_5.ipynb'
        elif HWN == 6:
            notebook_path = 'BUSA523_Shrikaran_HW_6.ipynb'
        elif HWN == 7:
            notebook_path = 'BUSA523_Shrikaran_HW_7.ipynb'
        elif HWN == 8:
            x = input("Enter Q1 or Q2: ")
            if x == "Q1":
                notebook_path = 'BUSA523_Shrikaran_HW_8_Q1.ipynb'
            elif x == "Q2":
                notebook_path = 'BUSA523_Shrikaran_HW_8_Q2.ipynb'
            else:
                print("Invalid input")
                continue
        elif HWN == 9:
            x = input("Enter Q1 or Q2: ")
            if x == "Q1":
                notebook_path = 'BUSA523_Shrikaran_HW_9_Q1.ipynb'
            elif x == "Q2":
                notebook_path = 'BUSA523_Shrikaran_HW_9_Q2.ipynb'
            else:
                print("Invalid input")
                continue
        elif HWN == 10:
            x = input("Enter a or b: ")
            if x == "a":
                notebook_path = 'BUSA523_Shrikaran_HW_11_Def_Class.ipynb'
            elif x == 'b':
                notebook_path = 'BUSA523_Shrikaran_HW_11.ipynb'
            else:
                print("Invalid input")
                continue

        
        opened_notebook = open_notebook(notebook_path)
        
        markdown_content = notebook_to_markdown(opened_notebook)

        display_markdown(markdown_content)
    else:
        print("Invalid Input")
```
